```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n =100
x = np.random.rand(n)
y = np.random.rand(n)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x,y,facecolors='none',edgecolors='r',s=80,label='Data points')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
