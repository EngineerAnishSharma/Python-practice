```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n  =150
x = np.random.rand(n)
y = np.random.rand(n)
sizes = np.random.rand(n)*100
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x,y,c='r',linewidths=2,s=sizes,label='Ball')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
